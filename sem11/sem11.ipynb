```python
from sympy import *
```

```python
x = Symbol('x')
```

```python
f, g = Symbol('f(x)'), Symbol('g(x)')
```

```python
fx = x**3-50*x
```

```python
gx = -x**4+88*x**2-241
```

```python
func_all = fx - gx
```

```python
func_all
```

```python
plot(fx, gx)
```

```python
plot(func_all)
```

```python
q = solve(func_all)
```

```python
lst_x = [i.evalf(5, chop = True) for i in q]
```

```python
lst_x
```

```python
lst_y = [fx.subs(x, i) for i in lst_x]
```

```python
lst_y
```

```python
for i, (x, y) in enumerate(zip(lst_x, lst_y), start =1):
    print(f'  {f}  {g}- {i}: ({x}, {y})')
```
